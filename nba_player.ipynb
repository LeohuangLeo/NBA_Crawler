```python
from bs4 import BeautifulSoup
import requests
import time

start =time.time()

file_home =requests.get('https://www.basketball-reference.com/contracts/players.html')
html_home = BeautifulSoup(file_home.text, 'html.parser')
# num_players = html_home.find_all('div',{'class':'section_heading'})[0]
# num_players = num_players.find('h2').text
# num_player = ''
# for i in num_players:
#     if i.isdigit():
#         num_player += i
# num_player = int(num_player) - 1

ttl_player_name = []
ttl_player_data_per_pts = []
ttl_player_data_per_trb = []
ttl_player_data_per_ast = []
ttl_player_salary = []
num_player = 509

count = 20
for i in range(num_player):
    if i == count:
        continue
    elif i == count+1:
        count += 22
        continue
    else:
        player_ = html_home.find_all('tbody')[0]
        player_ = player_.find_all('tr')[i]
        player_ = player_.find_all('td')[0]
        player_ = player_.find_all('a')[0]
        player_link  = player_.get('href')
        file_data = 'https://www.basketball-reference.com'+player_link
        file_data = requests.get(file_data)
        html_data = BeautifulSoup(file_data.text, 'html.parser')
        player_name = html_data.find('h1').text
        ttl_player_name.append(player_name)
        print(player_name)
        
        for j in range(3,8,2):
            try:
                player_data = html_data.find_all('div', {'class':'p1'})[0].children
                player_data_1 = list(player_data)[j]
                player_data_per = player_data_1.find_all('p')[0].text
                if j == 3:
                    if player_data_per == '':
                        player_data_per = player_data_1.find_all('p')[1].text
                        ttl_player_data_per_pts.append(player_data_per)
                        print(player_data_per)
                    else:
                        ttl_player_data_per_pts.append(player_data_per)
                        print(player_data_per)
                elif j == 5:
                    if player_data_per == '':
                        player_data_per = player_data_1.find_all('p')[1].text
                        ttl_player_data_per_trb.append(player_data_per)
                        print(player_data_per)
                    else:
                        ttl_player_data_per_trb.append(player_data_per)
                        print(player_data_per)
                elif j == 7:
                    if player_data_per == '':
                        player_data_per = player_data_1.find_all('p')[1].text
                        ttl_player_data_per_ast.append(player_data_per)
                        print(player_data_per)
                    else:
                        ttl_player_data_per_ast.append(player_data_per)
                        print(player_data_per)
            except:
                if j == 3:
                    ttl_player_data_per_pts.append('NAN')
                    print('NAN')
                elif j == 5:
                    ttl_player_data_per_trb.append('NAN')
                    print('NAN')
                elif j == 7:
                    ttl_player_data_per_ast.append('NAN')
                    print('NAN')
        
for i in html_home.find_all('td',{'data-stat':'y1'}):
    player_salary = i.text
    player_salary = player_salary[1:].replace(',','')
    ttl_player_salary.append(player_salary)


        

print(ttl_player_name)
print(len(ttl_player_name))
print(ttl_player_data_per_pts)
print(len(ttl_player_data_per_pts))
print(ttl_player_data_per_trb)
print(len(ttl_player_data_per_trb))
print(ttl_player_data_per_ast)
print(len(ttl_player_data_per_ast))
print(ttl_player_salary)
print(len(ttl_player_salary))

end = time.time()
running_time = end-start
print("time cost: {:.5f} sec".format(running_time))
```
